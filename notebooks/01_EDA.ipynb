```python
# ============================================================
# FEATURE ENGINEERING: Original Data → ML Ready Dataset
# ============================================================
import pandas as pd
import numpy as np

# ── 1. Load ──────────────────────────────────────────────────
df = pd.read_excel('data/2024.xlsx')
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Date']   = pd.to_datetime(df['Date'], errors='coerce')
df['Month']  = df['Date'].dt.month

# ── 2. Test count per row ─────────────────────────────────────
def count_tests(t):
    if pd.isna(t): return 0
    return len([x for x in str(t).replace(' ', '').split('+') if x])

df['test_count'] = df['Tests'].apply(count_tests)

# ── 3. Test category keywords ─────────────────────────────────
BLOOD = ['CP','CBC','HBA1C','BSF','BSR','LFT','RFT','LIPID','ALT','ESR',
         'CRP','TSH','THYROID','CREATININE','UREA','BUN','ELECTROLYTES',
         'CALCIUM','VITAMIN','FERRITIN','IRON','HCV','HBSAG','MP','TYPHIDOT',
         'B-HCG','FSH','LH','PROLACTIN','AMH','PTH','URIC ACID','CHOLESTEROL',
         'TRIGLYCERIDE','ALBUMIN','BILIRUBIN','PT','APTT','SODIUM','POTASSIUM',
         'B-12','FOLIC','PHOSPHOROUS','MAGNESIUM','BLOOD GROUP','BLOOD GROOP',
         'HB','ANTI TPO','MICROALBUMIN','TIBC','R.A FACTOR','ICT MP']
RADIO = ['X-RAY','CT SCAN','MRI','U/S','ECHO','DOPPLER','OPG','ETT']
URINE = ['URINE R/E','URINE C/S','ALBUMIN CREATININE']
STOOL = ['STOOL R/E','STOOL C/S','H.PYLORI']
SPUTUM= ['SPUTUM','AFB']

def cat_count(t, keywords):
    if pd.isna(t): return 0
    t_up = str(t).upper()
    return sum(1 for kw in keywords if kw in t_up)

for col, kws in [('b', BLOOD), ('r', RADIO), ('u', URINE), ('s', STOOL), ('sp', SPUTUM)]:
    df[col] = df['Tests'].apply(lambda t: cat_count(t, kws))

# ── 4. Patient-level aggregation ──────────────────────────────
def get_trend(months):
    if len(months) < 2: return 'Stable'
    # tests per month
    return 'Stable'  # placeholder — calculated properly below

grp = df.groupby("Patient,s Name")

# Tests per month per patient (for trend)
monthly = (df.groupby(["Patient,s Name", "Month"])['test_count']
             .sum()
             .reset_index())

def compute_trend(name):
    pat = monthly[monthly["Patient,s Name"] == name].sort_values('Month')
    if len(pat) < 2:
        return 'Stable'
    first_half = pat.iloc[:len(pat)//2]['test_count'].mean()
    second_half = pat.iloc[len(pat)//2:]['test_count'].mean()
    diff = second_half - first_half
    if diff > 1:   return 'Increasing'
    if diff < -1:  return 'Decreasing'
    return 'Stable'

patient_names = df["Patient,s Name"].unique()
trend_map = {name: compute_trend(name) for name in patient_names}

# Aggregate
ml = grp.agg(
    Gender            = ('Gender', 'first'),
    Visit_Count       = ('Date', 'nunique'),
    Test_Freq         = ('test_count', 'sum'),
    Total_Amount      = ('Amount', 'sum'),
    num_BLOOD_tests   = ('b',  'sum'),
    num_RADIOLOGY_tests=('r', 'sum'),
    num_URINE_tests   = ('u',  'sum'),
    num_STOOL_tests   = ('s',  'sum'),
    num_SPUTUM_tests  = ('sp', 'sum'),
).reset_index()

# ── 5. Remaining features ──────────────────────────────────────
# Test diversity — unique test names per patient
def unique_tests(name):
    tests = df[df["Patient,s Name"] == name]['Tests'].dropna()
    all_t = []
    for t in tests:
        all_t += [x.strip().upper() for x in str(t).split('+') if x.strip()]
    return len(set(all_t))

ml['Test_Diversity']    = ml["Patient,s Name"].apply(unique_tests)
ml['Repeat_Test_Count'] = (ml['Test_Freq'] - ml['Test_Diversity']).clip(lower=0)
ml['Monthly_Test_Trend']= ml["Patient,s Name"].map(trend_map)
ml['Gender']            = ml['Gender'].map({'M': 1, 'F': 0})

# ── 6. Moral Hazard Index (0–3 scale) ─────────────────────────
repeat_thresh = ml['Repeat_Test_Count'].quantile(0.75)
amount_thresh  = ml['Total_Amount'].quantile(0.75)
freq_thresh    = ml['Test_Freq'].quantile(0.75)

ml['MHI_repeat'] = (ml['Repeat_Test_Count'] >= repeat_thresh).astype(int)
ml['MHI_amount'] = (ml['Total_Amount']       >= amount_thresh).astype(int)
ml['MHI_freq']   = (ml['Test_Freq']          >= freq_thresh).astype(int)
ml['Moral_Hazard_Index'] = ml['MHI_repeat'] + ml['MHI_amount'] + ml['MHI_freq']

# ── 7. Target variable ────────────────────────────────────────
threshold = ml['Total_Amount'].quantile(0.80)
print(f"✅ High Cost Utilizer threshold (top 20%): Rs. {threshold:,.0f}")
ml['High_Cost_Utilizer'] = (ml['Total_Amount'] >= threshold).astype(int)

# ── 8. Final cleanup ──────────────────────────────────────────
final_cols = ['Gender','Visit_Count','Test_Freq','Test_Diversity',
              'Repeat_Test_Count','num_BLOOD_tests','num_RADIOLOGY_tests',
              'num_URINE_tests','num_STOOL_tests','num_SPUTUM_tests',
              'Monthly_Test_Trend','Moral_Hazard_Index','High_Cost_Utilizer']

ml_final = ml[final_cols].copy()
ml_final = ml_final.drop_duplicates()

# ── 9. Summary ────────────────────────────────────────────────
print(f"\n📊 Final ML Dataset Shape: {ml_final.shape}")
print(f"\n🎯 Target Distribution:")
print(ml_final['High_Cost_Utilizer'].value_counts())
print(f"\n🔥 Moral Hazard Index Distribution:")
print(ml_final['Moral_Hazard_Index'].value_counts().sort_index())
print(f"\n📈 Monthly Test Trend:")
print(ml_final['Monthly_Test_Trend'].value_counts())
print(f"\n✅ Null values: {ml_final.isnull().sum().sum()}")
print("\n📋 Sample (5 rows):")
print(ml_final.head())

# ── 10. Save ──────────────────────────────────────────────────
ml_final.to_csv('ml_ready_health_data_v2.csv', index=False)
print("\n💾 Saved: ml_ready_health_data_v2.csv")
```

```python
# ============================================================
# STEP 6: Outlier Detection
# ============================================================

print("=" * 50)
print("📊 OUTLIER DETECTION — IQR Method")
print("=" * 50)

num_cols = ['Visit_Count', 'Test_Freq', 'Test_Diversity', 'Repeat_Test_Count',
            'num_BLOOD_tests', 'num_RADIOLOGY_tests', 'num_URINE_tests',
            'num_STOOL_tests', 'num_SPUTUM_tests', 'Moral_Hazard_Index']

# ── 1. Boxplots ───────────────────────────────────────────────
fig, axes = plt.subplots(2, 5, figsize=(18, 8))
fig.suptitle('Outlier Detection — Boxplot per Feature', fontsize=14, fontweight='bold')
axes = axes.flatten()

for i, col in enumerate(num_cols):
    ax = axes[i]
    bp = ax.boxplot(df[col], patch_artist=True, notch=False,
                    boxprops=dict(facecolor='steelblue', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2),
                    flierprops=dict(marker='o', markerfacecolor='tomato',
                                    markersize=4, alpha=0.5),
                    whiskerprops=dict(linewidth=1.5),
                    capprops=dict(linewidth=1.5))
    ax.set_title(col, fontweight='bold', fontsize=10)
    ax.set_ylabel('Value')
    ax.spines[['top', 'right']].set_visible(False)

    # IQR
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)][col]
    ax.text(0.97, 0.95, f'Outliers: {len(outliers)}',
            transform=ax.transAxes, ha='right', va='top',
            fontsize=9, color='darkred',
            bbox=dict(boxstyle='round,pad=0.2', fc='lightyellow', alpha=0.8))

plt.tight_layout()
plt.savefig('../results/outlier_boxplots.png', dpi=150, bbox_inches='tight')
plt.show()
print("💾 Saved: results/outlier_boxplots.png")

# ── 2. IQR Summary table ──────────────────────────────────────
print("\n📋 IQR Outlier Summary:")
print("-" * 65)
print(f"  {'Feature':<25} {'Q1':>6} {'Q3':>6} {'IQR':>6} {'Lower':>7} {'Upper':>7} {'Outliers':>9} {'%':>6}")
print("-" * 65)

total_outliers = 0
outlier_info = {}

for col in num_cols:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    n = len(outliers)
    pct = n / len(df) * 100
    total_outliers += n
    outlier_info[col] = {'lower': lower, 'upper': upper, 'count': n}
    flag = ' ⚠️' if pct > 5 else ''
    print(f"  {col:<25} {Q1:>6.1f} {Q3:>6.1f} {IQR:>6.1f} {lower:>7.1f} {upper:>7.1f} {n:>9} {pct:>5.1f}%{flag}")

print("-" * 65)
print(f"  {'Total outlier rows (unique)':<40}", end="")
all_outlier_idx = set()
for col in num_cols:
    info = outlier_info[col]
    idx = df[(df[col] < info['lower']) | (df[col] > info['upper'])].index
    all_outlier_idx.update(idx)
print(f"{len(all_outlier_idx):>6} ({len(all_outlier_idx)/len(df)*100:.1f}%)")

# ── 3. Decision ───────────────────────────────────────────────
print("\n📌 Decision:")
print("  → Outliers drop nahi karenge — yeh genuine high-cost patients hain")
print("  → ML models (Random Forest, SVM) outliers ko handle kar sakte hain")
print("  → Preprocessing mein RobustScaler use karenge jo outlier-resistant hai")

print("\n✅ Step 6 Complete!")
```

```python
# ============================================================
# STEP 7: Encoding & Preprocessing
# ============================================================
from sklearn.preprocessing import RobustScaler, LabelEncoder

print("=" * 50)
print("⚙️  ENCODING & PREPROCESSING")
print("=" * 50)

# ── 1. Copy original ──────────────────────────────────────────
df_processed = df.copy()

# ── 2. Label Encode — Monthly_Test_Trend ─────────────────────
print("\n📌 Step 7.1 — Label Encoding: Monthly_Test_Trend")
print("-" * 45)
trend_map = {'Decreasing': 0, 'Stable': 1, 'Increasing': 2}
df_processed['Monthly_Test_Trend'] = df_processed['Monthly_Test_Trend'].map(trend_map)

print("  Encoding map:")
for k, v in trend_map.items():
    count = df[df['Monthly_Test_Trend'] == k].shape[0]
    print(f"    {k:12} → {v}  (n={count})")

print(f"\n  After encoding unique values: {sorted(df_processed['Monthly_Test_Trend'].unique())}")
print("  ✅ Encoding complete!")

# ── 3. Feature & Target split ─────────────────────────────────
print("\n📌 Step 7.2 — Feature & Target Split")
print("-" * 45)

feature_cols = ['Gender', 'Visit_Count', 'Test_Freq', 'Test_Diversity',
                'Repeat_Test_Count', 'num_BLOOD_tests', 'num_RADIOLOGY_tests',
                'num_URINE_tests', 'num_STOOL_tests', 'num_SPUTUM_tests',
                'Monthly_Test_Trend', 'Moral_Hazard_Index']

X = df_processed[feature_cols]
y = df_processed['High_Cost_Utilizer']

print(f"  Features (X): {X.shape}")
print(f"  Target  (y): {y.shape}")
print(f"  Feature columns: {list(X.columns)}")

# ── 4. RobustScaler ───────────────────────────────────────────
print("\n📌 Step 7.3 — RobustScaler (outlier-resistant)")
print("-" * 45)

scaler = RobustScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X),
                         columns=feature_cols)

print("  Before scaling (Test_Freq):")
print(f"    min={X['Test_Freq'].min()}, max={X['Test_Freq'].max()}, mean={X['Test_Freq'].mean():.2f}")
print("  After scaling (Test_Freq):")
print(f"    min={X_scaled['Test_Freq'].min():.2f}, max={X_scaled['Test_Freq'].max():.2f}, mean={X_scaled['Test_Freq'].mean():.2f}")
print("  ✅ Scaling complete!")

# ── 5. Scaling visualization ──────────────────────────────────
fig, axes = plt.subplots(2, 6, figsize=(18, 8))
fig.suptitle('Before vs After RobustScaler', fontsize=14, fontweight='bold')

for i, col in enumerate(feature_cols):
    # Before
    axes[0, i % 6].hist(X[col], bins=20, color='steelblue',
                         edgecolor='black', linewidth=0.4, alpha=0.8)
    axes[0, i % 6].set_title(f'{col}\n(before)', fontsize=8, fontweight='bold')
    axes[0, i % 6].spines[['top', 'right']].set_visible(False)

    # After
    axes[1, i % 6].hist(X_scaled[col], bins=20, color='tomato',
                          edgecolor='black', linewidth=0.4, alpha=0.8)
    axes[1, i % 6].set_title(f'{col}\n(after)', fontsize=8, fontweight='bold')
    axes[1, i % 6].spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.savefig('../results/scaling_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n💾 Saved: results/scaling_comparison.png")

# ── 6. Final processed dataframe ──────────────────────────────
print("\n📌 Step 7.4 — Final Processed Dataset")
print("-" * 45)
df_final = X_scaled.copy()
df_final['High_Cost_Utilizer'] = y.values

print(f"  Shape       : {df_final.shape}")
print(f"  Null values : {df_final.isnull().sum().sum()}")
print(f"  Dtypes      :")
print(df_final.dtypes)
print("\n  Sample (3 rows):")
print(df_final.head(3).round(3))

print("\n" + "=" * 50)
print("✅ Step 7 Complete!")
print("=" * 50)
```
